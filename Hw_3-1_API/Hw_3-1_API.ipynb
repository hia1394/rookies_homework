```python
import requests
import pandas as pd
import json
from pprint import pprint 

# 네이버 API 호출을 위한 공통 함수

def search_books(query):
    payload = {
        'query': query,
        'display': 50,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/book.json'

    headers = {
        'X-Naver-Client-Id': 'xkrBm3fGKPpXn9RulMkf',  # 네이버 API 클라이언트 ID
        'X-Naver-Client-Secret': '2ZMD4TeKED'  # 네이버 API 클라이언트 시크릿
    }
        
    res = requests.get(url, params=payload, headers=headers)
    if res.status_code == 200:
        return res.json()['items']


query = "파이썬"
#질문 1. 검색어로 찾은 책 목록을 json 파일로 저장하기
sea_book = search_books(query)

with open('./books.json', 'w', encoding='utf-8') as f:
    json.dump(sea_book, f, ensure_ascii=False, indent=4)

#질문 2. books.json 파일을 Pandas DataFrame으로 저장하기
with open('./books.json', 'r', encoding='utf-8') as f:
        book_data = json.load(f) 
    
df_books = pd.DataFrame(book_data) 

#질문 3. 검색어로 찾은 책 목록 출력하기
df_books

# 질문4. 가격이 2만원 이상인 책만 출력하기
df_books.loc[df_books['discount'] >= '20000',['title','author','discount','publisher','pubdate']]\
   .sort_values(by='discount',ascending=False).reset_index(drop=True)


# 질문 6. 출판사가 인피니티북스인 책만 출력하기
df_books.loc[df_books['publisher']=='인피니티북스',['title','link','author','discount','publisher','pubdate','isbn']]\
    .reset_index(drop=True)
```
